```python
# import dependencies
import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
import tensorflow as tf
import sklearn as skl
```

```python
#  Import and read the csv.
data_df = pd.read_csv("Triangle_role/Resources/sample_data.csv")
data_df.head()
```

```python
data_df = data_df.drop(columns=["C_S"])
data_df.head()
```

```python
# creating the scaler instance
scaler = StandardScaler()

# Fitting the scaler
scaled_df = scaler.fit_transform(data_df)
```

```python
scaled_df = pd.DataFrame(scaled_df, columns=data_df.columns)
scaled_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = data_df["Cost"].values
X = data_df.drop(["Cost"],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=22)
```

```python
# Fit the scaler
scaler.fit(X_train)

# Scale the data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Create scaler instance
X_scaler = skl.preprocessing.StandardScaler()

# Fit the scaler
X_scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create the Keras Sequential model
nn_model = tf.keras.models.Sequential()
```

```python
# Add our first Dense layer, including the input layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="relu", input_dim=2))
```

```python
# Add the output layer that uses a probability activation function
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
# Check the structure of the Sequential model
nn_model.summary()
```

```python
# Compile the Sequential model together and customize metrics
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Fit the model to the training data
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=25)
```
